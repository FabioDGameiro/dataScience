```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
 
ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("UF X valor - 2008/Ago")
plt.show()
```

```
dados["2020/Ago"] = pd.to_numeric(dados["2020/Ago"],errors='coerce')
print(dados["2020/Ago"].dtypes)

```

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

dados["2020/Ago"] = pd.to_numeric(dados["2020/Ago"],errors='coerce')
ax202008 = dados.plot(x="Unidade da Federação", y="2020/Ago", kind="bar", figsize=(9,6))
ax202008.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("UF X valor - 2020/Ago")
plt.show()
```
